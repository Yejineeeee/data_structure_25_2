- 수행하는 작업의 목록으로 데이터 타입을 나타낸 것

### 3. 파이썬 내장 리스트의 한계

- 파이썬 리스트는 배열로 구현되어 있음

**한계**

1. **원소 삽입 시, 삽입하는 자리 오른쪽부터 끝까지 모든 원소를 한 칸씩 시프트 해주는 부담이 존재함**
    
    → 3번 자리에 삽입하면서 원래 배열에서 3번 ~ 끝자리에 있던 모든 원소가 오른쪽으로 한 칸씩 이동하는 것을 볼 수 있음
    
    - 삭제에서는 반대 방향으로 시프트가 발생
2. **배열은 공간의 크기를 미리 정해서 받을 수밖에 없음**
    
    → 받아놓은 공간이 꽉 차 있는 상태에서 삽입이 시도되면 배열의 한계를 넘어가 삽입을 할 수밖에 없다.
    
    → **새 배열을 할당 받아 기존 배열의 내용을 새 배열로 복사하고, 리스트가 새 배열과 매핑되도록 함.**

## 03) 연결 리스트

### 1. 연결 리스트의 객체 구조

- **연결 리스트** : 배열의 공간 낭비를 피할 수 있는 자료 구조
- **동적 할당(dynamic allocation)** : 원소가 추가될 때마다 공간을 할당받아 추가하는 방식

- 연결 리스트의 첫 번째 노드에 대한 reference field : **__head**   
- 리스트에 들어 있는 원소의 총 수를 저장하는 field : **__numItems**   
- 두 필드 **__head, __numItems**는 사용자가 직접 접근하지 않으며,   
- 연결 리스트 개체의 메소드들이 작업할 때 필요함.   
- __ (언더바 2개)로 시작하는 필드나 메소드는 객체 내부에서만 사용함 

      -> **프라이빗 필드 or 프라이빗 메소드**   

```python
class ListNode:
    def __init__(self, newItem, nextNode: 'ListNode'):
        self.item = newItem
        self.next = nextNode
# a = ListNode(x, Node) -> 필드 item 값 = x,
# 필드 next 값 = None인 노드 객체 a를 생성
# nextNode:'ListNode')는 nextNode의 타입이 연결리스트임을 강변함.
```

#### 원소 삽입 위치에 따른 삽입하는 방법은 알고리즘 [5-1]과 같다.   
원소를 맨 앞에 삽입하는 경우와 그렇지 않은 경우로 나누어 처리한다.   
여기서 변수 i는 i번 원소로서 x를 삽입함을 뜻한다.

```python
# 예제 1

class Node:
    def __init__(self, item, nxt=None):
        self.item = item
        self.next = nxt

class LinkedList:
    def __init__(self):
        self.__head = None
        self.__numItems = 0

    def __getNode(self, idx):  # 0-based, idx번째 노드 반환
        cur = self.__head
        for _ in range(idx):
            if cur is None:
                raise IndexError("index out of range")
            cur = cur.next
        return cur

    def insert(self, i, x):    # 0 ≤ i ≤ len
        if not (0 <= i <= self.__numItems):
            raise IndexError("index out of range")

        newNode = Node(x)
        if i == 0:
            # newNode.item = x  # Node 생성자에서 이미 설정
            newNode.next = self.__head
            self.__head = newNode
        else:
            prev = self.__getNode(i - 1)
            newNode.next = prev.next
            prev.next = newNode
        self.__numItems += 1

    def to_list(self):
        out, cur = [], self.__head
        while cur:
            out.append(cur.item)
            cur = cur.next
        return out

# 사용 예
L = LinkedList()
L.insert(0, "A")
L.insert(1, "B")
L.insert(1, "X")  # A, X, B
print(L.to_list())  # ['A', 'X', 'B']
```
