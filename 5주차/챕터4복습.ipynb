### 5) 복사호출, 참조호출, 할당호출

- 복사 호출은 변수 x의 값을 넘김
- 참조 호출은 변수 x의 레퍼런스 (식별자, 포인터 또는 주소)를 넘김

- 위치 인수 : 함수 호출 시 인수의 순서에 따라 매개변수와 대응되는 인수

#### Unpacking 함수 **(*리스트)**
- Packing (패킹) : 인자로 받은 여러 개의 값을 하나의 객체로 합쳐서 받을 수 있도록 하는 것   
-> **위치인자 패킹** : * 한 개를 매개 변수 앞에 붙임

- Unpacking (언패킹) : 여러 개의 객체를 포함하고 있는 하나의 객체를 풀어준다.   
함수에서 언패킹을 할 때는, 매개 변수에서 * 를 붙이는 게 아니라 **인자 앞에 * 를 붙여 사용한다.**

```python
# 예시 1
def print_no(a, b, c) :
    print(a)
    print(b)
    print(c)
    
print_no(10, 20, 30)

print("\n")

# 예시 2
x = [10, 20, 30]
print_no(*x)

print("\n")

# 예시 3
print_no(*[20, 30, 40])

print("\n")
```

#### Unpacking 함수 (*튜플)
-> 튜플 : 바꿀 수 없음

```python
fruits = ("apple", "banana", "cherry")

(green, yellow, red) = fruits
print(green)

fruit = ("apple", "banana", "cherry", "strawberry", "raspberry")
(green, yellow, *red) = fruit

print("\n")

print(green)

print(yellow)

print(red)

print(*red)
```

#### 가변 인수(variable argument) 함수
def 함수 이름(*매개변수) :
    코드

-> 가변 인수 함수는 매개변수 앞에 *를 붙임

*args
-> 매개 변수의 길이를 정확히 모를 때

```python
def prt_no(*args):
    for arg in args:
        print(arg)
        
prt_no(10)

print("\n")

prt_no(10, 20, 30, 40)

print("\n")

y = [100, 200, 300]
prt_no(*y)
```

#### 키워드 인수
: 함수에 인수를 넣을 때 값이나 변수를 넣었으나, 아래는 인수의 순서를 기억하지 않아도 되는 키워드 인수라는 기능을 제공

```python
def personal_info(name, age, address):
    print('이름', name)
    print('나이', age)
    print('주소', address)
    
personal_info('홍길동', 25, '용인시 처인구 명지로 116')

print("\n")

# personal_info 함수를 키워드 인수 방식으로 호출 --> 함수 (키워드==값)
personal_info(name='성춘향', age=23, address='용인시 처인구 명지로')
```

#### 키워드 인수와 dictionary unpacking

```python
def personal_info(name, age, address):
    print('이름', name)
    print('나이', age)
    print('주소', address)
    
x = {'name':'홍길동', 'age':30, 'address':'용인시 처인구'}
personal_info(**x)

print("\n")

# dictionary는 * 대신 ** 를 사용 (두 번 Unpacking 해야 함)
x = {'name':'홍길동', 'age':30, 'address':'용인시 처인구'}
personal_info(*x)
```

#### 키워드 인수를 사용하는 가변 인수 함수

```python
# 키워드 인수를 사용하는 가변 인수 함수
# def 함수 이름 (**매개변수) :
#     코드

def personal_info(**kwargs) :
    for kw, arg in kwargs.items():
        print(kw, ':', arg, sep=' ')
        
personal_info(name="James")

print("\n")

x = {'name':'Susan'}
personal_info(**x)
```

#### 클래스 계수기 (Counter)

```python
class Counter:
    def __init__(self):     # 생성자
        self.count = 0      # : 무조건 실행해줘야 함.
    def increment(self):
        self.count += 1
    def get(self):
        return self.count
```

```python
b = Counter()
b.increment()
b.increment()
b.increment()
print(b.get())
```

```python
class Bank_Account:
    def __init__(self):
        self.balance = 0
    def deposit(self, amount):
        self.balance += amount
        print("\n amount deposited: ", amount)  # 저금
        return self.balance
    
    def withdraw(self, amount):
        self.balance -= amount
        print("\n amount withdraw: ", amount)  # 출금
        return self.balance
    
    def display(self):
        print("\n Net available balance: ", self.balance)   # 잔액 확인
        
a = Bank_Account()
a.deposit(100)
a.withdraw(40)
a.display()
```

비공개 속성(private attribute) : class 밖에서 비공개 속성에 접근하면 에러 발생

```python
class Person:
    def __init__(self, name, age, address, wallet):
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet      # 비공개 속성, 접근 불가
        
esther = Person("에스터", 34, "SF California", 10000)
esther.__wallet -= 10000
```

```python
class Person:
    def __init__(self, name, age, address, wallet):
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet      # 비공개 속성, 접근 불가
        
    def pay(self, amount):
        self.__wallet -= amount
        print('이제 {0}원이 남았네요.'.format(self.__wallet))   # 클래스 안의 메소드
        
esther = Person("에스터", 34, "SF California", 10000)
esther.pay(6000)
```

### 마지막 문단의 의미

파이썬의 관용구 if __name__ == "__main__": 를 변형한 부분   

파이썬이 모듈을 직접 실행할 때 그 파일의 전역 변수 __name__을 "__main__"으로 설정한다.   
( ex) if __name__ = "__main__": ) 

반대로 다른 파일이 import 해서 불러 쓰면 __name__은 모듈 이름(파일명)    
따라서 해당 파일은, __name__이 **정확히 "sample"**일 때만 실행된다.

즉, 파일 이름이 **sample.py**이고 그 파일을 다른 곳에서 import sample 하면,    
그 모듈 내부의 __name__ 값이 "sample"가 되어 실행된다.
